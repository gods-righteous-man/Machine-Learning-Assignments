# (g) Is there evidence of association of interactions of predictors with the response? To answer this question, run a full linear regression model with all pairwise interaction terms and state whether any interaction terms are statistically significant.

```python
import statsmodels.api as sm

X = ind_var.values
Y = dep_var.values

columns = ['AT', 'V', 'AP', 'RH']

A = []
for i in range(len(columns)):
    temp = X[:, i]
    A.append(temp)
    
print(A)
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        temp = X[:, i]*X[:, j]
        A.append(temp)
        

print(A)
# Finding the transpose of array A
A = np.array(A).T

temp = sm.add_constant(A)
model = sm.OLS(Y,temp)
fitting = model.fit()
details = fitting.summary()

print(details)





```

# Here it is observed that AT-V, AT-RH, V-AP and AP-RH have smaller p values. Therefore we can say that they are statistically significant.

# (i) KNN Regression: i. Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k  {1,2,...,100} that gives you the best fit. Plot the train and test errors in terms of 1/k.


```python
X = ind_var.values
Y = dep_var.values

#using normalizer
N = Normalizer().fit(X)
X_normalize = N.transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state =0)
anorm, bnorm, cnorm, dnorm = train_test_split(X_normalize, Y, test_size = 0.3, random_state = 0)
trainingerror = []
testingerror = []
ntrainerror = []
ntesterror = []

error = float('inf')
norm_error = float('inf')

for k in range(1, 101):
    A = KNeighborsRegressor(n_neighbors = k).fit(X_train, Y_train)
    B = KNeighborsRegressor(n_neighbors = k).fit(anorm, cnorm)
    Ptrained = A.predict(X_train)
    Ptested = A.predict(X_test)
    Ptrained_norm = B.predict(anorm)
    Ptested_norm = B.predict(bnorm)
    
    trainingerror.append(mean_squared_error(Y_train, Ptrained))
    testingerror.append(mean_squared_error(Y_test, Ptested))
    ntrainerror.append(mean_squared_error(cnorm, Ptrained_norm))
    ntesterror.append(mean_squared_error(dnorm, Ptested_norm))
    if min(error, mean_squared_error(Y_test, Ptested)) == mean_squared_error(Y_test, Ptested):
        val1 = k
        error = min(error, mean_squared_error(Y_test, Ptested))
    
    if min(norm_error, mean_squared_error(dnorm, Ptested_norm)) == mean_squared_error(dnorm, Ptested_norm):
        val2 = k
        norm_error = min(norm_error, mean_squared_error(dnorm, Ptested_norm))
        
plt.figure(figsize=(8, 5))
plt.title("K Nearest Neighbor Graph")
plt.plot(1 / np.arange(1,101), trainingerror, color='red', label='Training error')
plt.plot(1 / np.arange(1,101), testingerror, color='blue', label='Test error')
plt.plot(1 / np.arange(1,101), ntrainerror, color='green', label='Normal train error')
plt.plot(1 / np.arange(1,101), ntesterror, color='orange', label='Normal test error')
plt.legend()
plt.xlabel('1 / Number of neighbors (K Value)')
plt.ylabel('Mean Squared Errors')

# plt.show()
print(len(trainingerror))
print(len(testingerror))
print(len(ntrainerror))
print(len(ntesterror))

print("The best fit is obtained for raw data at K = ", val1)
print("And its min error is :", error)
print("The best fit is obtained for raw data at k = ", val2)
print("And its min error is :", norm_error)

    
```
