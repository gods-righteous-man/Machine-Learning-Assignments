# iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate, 3 which is obtained by some value of k, against the size of training set, when the size of training set is N ∈ {10, 20, 30, . . . , 210}. 4 Note: for each N, select your training set by choosing the first bN/3c rows of Class 0 and the first N − bN/3c rows of Class 1 in the training set you created in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from {1, 6, 11, . . . , 196}. This plot is called a Learning Curve.

```python
Ndict = {}
for i in range(10, 211, 10):
    A = Data_ind_var_train[0:i - int(i/3)]
    B = Data_ind_var_train[140: 140 + int(i/3)]
    frames = [A, B]
    M = pd.concat(frames, ignore_index = True)
    C = Data_dep_var_train[0: i - int(i/3)]
    D = Data_dep_var_train[140: 140 + int(i/3)]
    frames = [C, D]
    N = pd.concat(frames, ignore_index = True)
    edict = {}
    for k in range(1, i+1, 5):
        c = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
        c.fit(M, N.values.ravel())
        pred = c.predict(Data_ind_var_test)
        acc = accuracy_score(Data_dep_var_test, pred)
        e = 1 - acc
        edict[k] = e
    Ndict[i] = edict

# print(Ndict)
#finding minimum value of error 
optimum_error = []
for i in Ndict.values():
    print(min(i.values()))
    optimum_error.append(min(i.values()))
    
```

```python
x = []
for i in range(10, 211, 10):
    x.append(i)
y = optimum_error



mplt.figure(figsize=(15,8))
mplt.plot(x, y, color = 'red', label = 'test error rate')
mplt.xlabel('Values of N')
mplt.ylabel('Minimum Errors')
mplt.legend()
mplt.show()
```

# (d) Replace the Euclidean metric with the following metrics5 and test them. Summarize the test errors (i.e., when k = k∗) in a table. Use all of your training data and select the best k when {1, 6, 11, . . . , 196}.

# i. Minkowski Distance:

# A. which becomes Manhattan Distance with p = 1.

```python
#Manhattan distance


minkow = {}
for k in range(1,196,5):
    #generating and fitting model
    c = KNeighborsClassifier(metric = 'minkowski', n_neighbors = k, p = 1)
    c.fit(Data_ind_var_train, Data_dep_var_train.values.ravel())
    #calculating predicted values on test data and checking error rate
    predicted_val_test_data = c.predict(Data_ind_var_test)
    acc_testing = accuracy_score(Data_dep_var_test, predicted_val_test_data )
    e_testing = 1 - acc_testing
    minkow[k] = e_testing

print(minkow)

a = list(minkow.keys())
b = list(minkow.values())

position = b.index(min(b))
best_k = a[position]
print("Best value of k: ", best_k)

manh_k = best_k
man_err = b[position]
```

```python
x = a
y = b

mplt.figure(figsize=(15,8))
mplt.plot(x, y, color = 'red', label = 'test error rate')
mplt.xlabel('Values of K')
mplt.ylabel('Errors')
mplt.legend()
mplt.show()
```

```python
#chebishev
#place float('inf') in place of p to find infinite value



chebi = {}
for k in range(1,196,5):
    #generating and fitting model
    c = KNeighborsClassifier(metric = 'minkowski', n_neighbors = k, p = float('inf'))
    c.fit(Data_ind_var_train, Data_dep_var_train.values.ravel())
    #calculating predicted values on test data and checking error rate
    predicted_val_test_data = c.predict(Data_ind_var_test)
    acc_testing = accuracy_score(Data_dep_var_test, predicted_val_test_data )
    e_testing = 1 - acc_testing
    chebi[k] = e_testing

print(chebi)

a = list(chebi.keys())
b = list(chebi.values())

position = b.index(min(b))
best_k = a[position]
print("Best value of k: ", best_k)
print("Error: ", b[position])

chebi_k = best_k
chebi_err = b[position]
```

```python
x = a
y = b

mplt.figure(figsize=(15,8))
mplt.plot(x, y, color = 'red', label = 'test error rate')
mplt.xlabel('Values of K')
mplt.ylabel('Errors')
mplt.legend()
mplt.show()
```

# ii. Mahalanobis Distance.6

```python
#mahalanobis

maha = {}
for k in range(1,196,5):
    #generating and fitting model
    c = KNeighborsClassifier(metric = 'mahalanobis', n_neighbors = k, metric_params = {'VI': np.cov(Data_ind_var_train)}, algorithm = 'brute')
    c.fit(Data_ind_var_train, Data_dep_var_train.values.ravel())
    #calculating predicted values on test data and checking error rate
    predicted_val_test_data = c.predict(Data_ind_var_test)
    acc_testing = accuracy_score(Data_dep_var_test, predicted_val_test_data )
    e_testing = 1 - acc_testing
    maha[k] = e_testing

print(maha)

a = list(maha.keys())
b = list(maha.values())

position = b.index(min(b))
best_k = a[position]
print("Best value of k: ", best_k)
print("Error: ", b[position])

maha_k = best_k
maha_err = b[position]
```

```python
x = a
y = b

mplt.figure(figsize=(15,8))
mplt.plot(x, y, color = 'red', label = 'test error rate')
mplt.xlabel('Values of K')
mplt.ylabel('Errors')
mplt.legend()
mplt.show()
```

# (e) The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k ∈ {1, 6, 11, 16, . . . , 196}.

```python
#weights euclidean




minkow = {}
for k in range(1,201,5):
    #generating and fitting model
    c = KNeighborsClassifier(metric = 'minkowski', n_neighbors = k, p = 2, weights = 'distance')
    c.fit(Data_ind_var_train, Data_dep_var_train.values.ravel())
    #calculating predicted values on test data and checking error rate
    predicted_val_test_data = c.predict(Data_ind_var_test)
    acc_testing = accuracy_score(Data_dep_var_test, predicted_val_test_data )
    e_testing = 1 - acc_testing
    minkow[k] = e_testing

print(minkow)

a = list(minkow.keys())
b = list(minkow.values())

position = b.index(min(b))
best_k = a[position]
print("Best value of k: ", best_k)

manh_kecl = best_k
man_errecl = b[position]
```
