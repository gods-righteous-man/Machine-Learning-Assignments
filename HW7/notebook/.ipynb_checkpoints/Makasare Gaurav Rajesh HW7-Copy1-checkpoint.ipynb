```python
X = df.iloc[:, :22]
y = df.iloc[:, 22:25]
c = y.columns
d = []
l = []
```

```python
#for performing procedure 50 times
for i in range(0, 50):
    predscore = []
    score = {}
    val_label = []
    print()
    print(i+1, ': \n',)
    k = 2
    while k < 51:
        kmmodel = KMeans(n_clusters=k, random_state=k)
        kmlabels = kmmodel.fit_predict(X)
        score[silhouette_score(X, kmlabels)] = k
        val_label.append(kmlabels)
        predscore.append(silhouette_score(X, kmlabels))
        k += 1
    t = score[max(list(score.keys()))]
    print("Optimum k: ", t)
    y['pred_labels'] = val_label[predscore.index(max(predscore))]
    #print(temp)
    print("Majority label are as follows:")
    for m in list(set(y['pred_labels'])):
        t2 = y[y['pred_labels'] == m]
        print("\n")
        print("Cluster: ", m)
        for p in c:
            t3 = dict()
            for q in list(set(t2[p])):
                t3[list(t2[p]).count(q)] = q
            maxlabel = t3[max(t3.keys())]
            print(p , " : ", maxlabel)
    for p in c:
        temp = 'new ' + str(p)
        a = y['pred_labels']
        b = a.map(y.groupby('pred_labels')[p].agg(pd.Series.mode))
        y[temp] = b
        d.append(hamming(y[p], y[temp]))
        l.append(hamming_loss(y[p], y[temp]))
            
            
        
    
```

```python
d1 = np.average(d)
l1 = np.average(l)
print("Hamming distance: ", d1)
print("Hamming score: ", l1)
print("Hamming loss: ", 1 - l1)
```
